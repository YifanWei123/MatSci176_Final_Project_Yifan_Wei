```python
import ase.io
from pathlib import Path

TRAIN_MD = "/home/jovyan/ML_HEPO/diff_com_T_OV/npt_final/train_fps.extxyz"
TRAIN_PERTURB = "/home/jovyan/ML_HEPO/random_variation/prepared_dataset_split/train_perturb.extxyz"

# VAL_MD = "/home/jovyan/ML_HEPO/diff_com_T_OV/npt_final/val_ood_r00.extxyz"
# VAL_PERTURB = "/home/jovyan/ML_HEPO/random_variation/prepared_dataset_split/val_perturb.extxyz"

OUT_DIR = "/home/jovyan/ML_HEPO/random_variation/prepared_dataset_final"
Path(OUT_DIR).mkdir(parents=True, exist_ok=True)

OUT_TRAIN = f"{OUT_DIR}/train_mixed.extxyz"
# OUT_VAL   = f"{OUT_DIR}/val_mixed.extxyz"

for f in [OUT_TRAIN]:
    Path(f).unlink(missing_ok=True)

def read_all(path):
    return list(ase.io.iread(path, index=":"))

def add_source(frames, source):
    for a in frames:
        a.info["source"] = source
    return frames

train_md = add_source(read_all(TRAIN_MD), "md")
train_perturb = add_source(read_all(TRAIN_PERTURB), "perturb")
# val_md = add_source(read_all(VAL_MD), "md")
# val_perturb = add_source(read_all(VAL_PERTURB), "perturb")

train_mixed = train_md + train_perturb
# val_mixed   = val_md + val_perturb

print("train_md      =", len(train_md))
print("train_perturb =", len(train_perturb))
print("train_mixed   =", len(train_mixed))
# print("val_md        =", len(val_md))
# print("val_perturb   =", len(val_perturb))
# print("val_mixed     =", len(val_mixed))

ase.io.write(OUT_TRAIN, train_mixed)
# ase.io.write(OUT_VAL, val_mixed)

print("\nWrote:")
print(OUT_TRAIN)
# print(OUT_VAL)
```
